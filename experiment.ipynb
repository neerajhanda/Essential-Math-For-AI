```python
import numpy as np
import pandas as pd
import seaborn
```

```python
N = 10000
```

```python
random_numbers = np.random.randn(N)
random_numbers

```

```python
x_axis=np.linspace(0,100, num=N)
```

```python
d={'X':x_axis, 'Y':random_numbers}
df = pd.DataFrame(d)
```

```python
plot1 = seaborn.jointplot(data=df,x='X',y='Y', kind='reg', color='k')
```
